```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# set the filepath and load
fp = 'reg2016_ed50_g/reg2016_ed50_g.shp'
#reading the file stored in variable fp
map_df = gpd.read_file(fp)
# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
map_df
```

```python
#plotting the map of the shape file preview of the maps without data in it
map_df.plot()
```

```python


choosed_regions = map_df.loc[map_df['regione'].isin(['Lazio', 'Toscana', 'Umbria', 'Molise', 'Campania'])]

title='Regiony włoskie zaopatrywane w wodę przez Acea Group'
```

```python
df = map_df.assign(is_studied='no')
df.loc[df['regione'].isin(['Lazio', 'Toscana', 'Umbria', 'Molise', 'Campania']), 'is_studied'] = 'yes'
```

```python
df.plot(column='is_studied', cmap='BuGn', figsize=(20,20), edgecolor='black')
```

```python
df.plot(column='is_studied', cmap='BuGn', figsize=(15,15), edgecolor='black')
plt.suptitle("Regiony włoskie zaopatrywane w wodę przez Acea Group", fontsize= 'xx-large', y=0.90)
```
